```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import sklearn.metrics as metrics
```

```python
AllData = np.zeros((210,120960))
for i in range(0,12):
    Loader = 'St'+ str(i+1) +'.npz'
    L = (np.load(Loader))
    AllData[:,(i)*10080:10080+((i)*10080)] = L
```

```python
TrainingDataComb1 = AllData[:,0:(3*(10080))]
TrainingDataComb2 = AllData[:,30240:60480]
TrainingDataComb3 = AllData[:,60480:90720]
TestDataComb1 = AllData[:,90720:120960]
```

```python
Labels = np.zeros((1,30240))
for i in range(0,10080):
    Labels[:,i] =  0
for i in range(10080,20160):
    Labels[:,i] =  1
for i in range(20160,30240):
    Labels[:,i] =  2
```

```python
np.shape(TrainingDataComb1.T)
```

```python
Labels
```

# Training

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(3)
])
```

```python
model.summary
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(TrainingDataComb1.T, Labels.T, epochs=105)
```

```python
model.fit(TrainingDataComb2.T, Labels.T, epochs=105)
```

```python
model.fit(TrainingDataComb3.T, Labels.T, epochs=105)
```

```python
test_loss, test_acc = model.evaluate(TestDataComb1.T, Labels.T, verbose=2)
```

# Prediction

```python
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
```

```python
predictions = probability_model.predict(TestDataComb1.T)
```

```python
Pred = np.zeros(30240)
for i in range(0,30240):
    Pred[i] = np.argmax(predictions[i])
```

```python
CM = metrics.confusion_matrix(Labels.T, Pred)
```

```python
CM
```

```python
CM = (CM/10080)*100
CM.round(2)
```

```python
plt.imshow(CM, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.show()
```
